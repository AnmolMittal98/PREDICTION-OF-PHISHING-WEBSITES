```python
X=df[['having_IP_Address', 'URL_Length', 'Shortining_Service',
       'having_At_Symbol', 'double_slash_redirecting', 'Prefix_Suffix',
       'having_Sub_Domain', 'SSLfinal_State', 'Domain_registeration_length',
       'Favicon', 'port', 'HTTPS_token', 'Request_URL', 'URL_of_Anchor',
       'Links_in_tags', 'SFH', 'Submitting_to_email', 'Abnormal_URL',
       'Redirect', 'on_mouseover', 'RightClick', 'popUpWidnow', 'Iframe',
       'age_of_domain', 'DNSRecord', 'web_traffic', 'Page_Rank',
       'Google_Index', 'Links_pointing_to_page', 'Statistical_report']]
y=df['Result']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.60, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree=DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
```

```python
predictions=dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
accuracy = 100.0 * accuracy_score(y_test,predictions)
print("The accuracy of your algorithm used on testing data is: " + str(accuracy)) 
```

```python
corr=df.corr()
plt.figure(figsize=(8,8))
# Transform it in a links data frame (3 columns only):
links = corr.stack().reset_index()
links.columns = ['having_IP_Address', 'URL_Length', 'Shortining_Service']
links
 
# Keep only correlation over a threshold and remove self correlation (cor(A,A)=1)
links_filtered=links.loc[ (links['Shortining_Service'] > 0.75) & (links['having_IP_Address'] != links['URL_Length']) ]
links_filtered
 
# Build your graph
G=nx.from_pandas_dataframe(links_filtered,'having_IP_Address', 'URL_Length')
 
# Plot the network:
nx.draw(G, with_labels=True, node_color='lightgreen', node_size=400, edge_color='orange', linewidths=1, font_size=12)
```
